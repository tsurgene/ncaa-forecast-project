```python
# Create Games dataframes:
MStats = pd.read_csv(DATA_DIR + 'MTeamStats.csv')
#WStats = pd.read_csv(DATA_DIR + 'WTeamStats.csv')
```

```python
# Show sample output
print('Mens:', MStats.shape)#, 'Womens:', WStats.shape)
print('Mens:', MStats.columns)
MStats.head(3)
```

```python
# Split 'Stats' dataframe (for subsequent build and matchups):

split_win = MStats.loc[:, ['Season', 'TeamID', 'NCAA', 'WinPCT', 'ConfPCT', 'WScore', 'LScore', 'MOV', 'SOS', 'OSRS', 'DSRS', 
                           'SRS', 'WTSPCT', 'WeFGPCT', 'WFGPCT', 'WFG3PCT', 'WFG3Ar', 'WFTPCT', 'WFTr', 'WORB', 'WTRB', 
                           'WAST', 'WTOV', 'WSTL', 'WBLK', 'WPF']].copy()
split_win['WIN'] = 1

split_loss = MStats.loc[:, ['Season', 'TeamID', 'NCAA', 'WinPCT', 'ConfPCT', 'WScore', 'LScore', 'MOV', 'SOS', 'OSRS', 'DSRS', 
                           'SRS', 'LTSPCT', 'LeFGPCT', 'LFGPCT', 'LFG3PCT', 'LFG3Ar', 'LFTPCT', 'LFTr', 'LORB', 'LTRB', 
                           'LAST', 'LTOV', 'LSTL', 'LBLK', 'LPF']].copy()
split_loss['WIN'] = 0

# Rename columns to prepare for concatenate dataframes
rename_win = {'WTSPCT':'TSPCT', 'WeFGPCT':'eFGPCT', 'WFGPCT':'FGPCT', 'WFG3PCT':'FG3PCT', 'WFG3Ar':'FG3Ar', 'WFTPCT':'FTPCT', 
              'WFTr':'FTr', 'WORB':'OR', 'WTRB':'TRB', 'WAST':'AST', 'WTOV':'TOV', 'WSTL':'STL', 'WBLK':'BLK', 'WPF':'PF'}

rename_loss = {'LTSPCT':'TSPCT', 'LeFGPCT':'eFGPCT', 'LFGPCT':'FGPCT', 'LFG3PCT':'FG3PCT', 'LFG3Ar':'FG3Ar', 'LFTPCT':'FTPCT', 
              'LFTr':'FTr', 'LORB':'OR', 'LTRB':'TRB', 'LAST':'AST', 'LTOV':'TOV', 'LSTL':'STL', 'LBLK':'BLK', 'LPF':'PF'}

split_win = split_win.rename(columns=rename_win)
split_loss = split_loss.rename(columns=rename_loss)

# Concatenate the two dataframes
MStats_df = pd.concat([split_win, split_loss])
MStats_df = MStats_df.sort_values(by=['Season', 'TeamID'])
# Show sample output
print('Mens:', MStats_df.shape)#, 'Womens:', WStats_df.shape)
print('Mens:', MStats_df.columns)
MStats_df.head(3)
```

```python
print('Missing Values:', MStats_df.isna().sum().sum())
#print('Missing Values:', WFavorites8_df.isna().sum().sum())
```

```python
# Impute missing values with mean based on TeamID over all seasons

# Group the DataFrame by the specific ID column to get mean
mean_id = MStats_df.groupby(['TeamID']).mean()

# Fill missing values with the mean value of each group
for col in mean_id.columns:
    MStats_df[col] = MStats_df[col].fillna(MStats_df[['TeamID']].merge(mean_id, on=['TeamID'], how='left')[col])




# Group the DataFrame by the specific ID column(s)
#MStats_df = MStats_df.groupby(['TeamID'])

# Calculate the mean of each group and fill missing values with it
#MStats_df = MStats_df.transform(lambda x: x.fillna(x.mean()))
```

```python
# Normalize selected columns using min-max range
def normalize_minmax(column):
    """
    Normalize a column to a min-max range.
    """
    min_val = min(column)
    max_val = max(column)
    normalized_col = [(x - min_val) / (max_val - min_val) for x in column]
    return normalized_col

# Select the columns you want to normalize
selected_cols = ['OR', 'DR', 'AST', 'TOV', 'STL', 'BLK', 'PF']

# Normalize the selected columns in your dataframe
for col in selected_cols:
    MStats2_df[col] = normalize_minmax(MStats2_df[col])

# Show sample output
print('Mens:', MStats2_df.shape)#, 'Womens:', womens_dropped.shape)
MStats2_df.head(3)
```

```python
# Standardize selected columns using standard scaler
def standardize_scaler(column):
    """
    Standardize a column using standard scaler.
    """
    mean_val = sum(column) / len(column)
    std_val = (sum([(x - mean_val)**2 for x in column]) / (len(column)-1))**0.5
    standardized_col = [(x - mean_val) / std_val for x in column]
    return standardized_col

# Select the columns you want to standardize
selected_cols = ['WScore', 'LScore', 'MOV', 'SOS', 'OSRS', 'DSRS', 'SRS', 'TSPCT']

# Standardize the selected columns in your dataframe
for col in selected_cols:
    MStats2_df[col] = standardize_scaler(MStats2_df[col])
    
# Show sample output
print('Mens:', MStats2_df.shape)#, 'Womens:', womens_dropped.shape)
MStats2_df.head(3)
```
