# Building Model

# SVM - Linear Kernel - 15 Features

```python
from sklearn import svm
from sklearn.model_selection import TimeSeriesSplit # will split data properly when we are doing feature selection
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier
#Create a svm Classifier
SVM = svm.SVC(kernel='linear') # Linear Kernel
split = TimeSeriesSplit(n_splits=3)
sfs = SequentialFeatureSelector(SVM, n_features_to_select=15, direction='forward', cv=split) #forward starts with zero features, pick the feature tha timporve model accuracy the most and then keep doing that until we get to 20
```

```python
removed_columns = ["season", "winner"] #columns that won't run in the model
```

```python
selected_columns = tournaments.columns[~tournaments.columns.isin(removed_columns)] #columns that we want to include in the model
```

```python
sfs.fit(tournaments[selected_columns], tournaments['winner'])
```

```python
predictors = list(selected_columns[sfs.get_support()])
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predictions = [] # list odf dataframe, each dataframe is the prediction for a single seasons
    seasons  = sorted(tournaments["season"].unique()) #list of seasons tha texist in data
    
    for i in range(start, len(seasons), step): #we progress with step parameter (one season at a time)
        season = seasons[i] # pull out the season number / 2003 will be the first season
        
        train = tournaments[tournaments["season"] < season] #for 2005, training will be 2004 and 2003
        test = tournaments[tournaments['season'] == season] #test will be 2005 (or whatever the current season is)
        
        model.fit(train[predictors], train['winner']) #this will take training data, all the predictors
        preds = model.predict(test[predictors]) #generate predictions from the test set we want 
        
        #by default model.fit  will return numpy array, but we want to convert it to a pandas series, keep index the same as test set index to make it easier to combin actuals and predictions
        preds = pd.Series(preds, index=test.index)
        
        combined = pd.concat([test['winner'], preds], axis=1) #axis=1 means treats target and preds as two columns, axis=0 would return one column and everything would be returned in rows
        combined.columns = ['actual', 'predictions'] #renaming columns
        
        all_predictions.append(combined) #this will be a list of five dataframes, one for each season, then we will combine all 
        
    return pd.concat(all_predictions)
```

```python
predictions = backtest(tournaments, SVM, predictors)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(predictions['actual'],predictions['predictions'])
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(predictions['actual'], predictions['predictions']))
```

## Manual Implementation of Measurement Metrics

To get a more granular look at the performance of the models used in this prediciton, I am calculating precision, recall and accuracy for each model to see how the model handles the data better. 

```python
predsDF = pd.merge(tournaments, predictions, left_index=True, right_index=True)
predsDF
# getting TP, TN, FP and FN value counts for the data
predsDF['truepos'] = np.where((predsDF['actual']==1) & (predsDF['predictions']==1),1,0)
predsDF['trueneg']= np.where((predsDF['actual']==0) & (predsDF['predictions']==0),1,0)
predsDF['falsepos']= np.where((predsDF['actual']==0) & (predsDF['predictions']==1),1,0)
predsDF['falseneg']= np.where((predsDF['actual']==1) & (predsDF['predictions']==0),1,0)
predsDF
predsDF2 = predsDF.groupby(['season']).sum()
predsDF2.columns
predsDF3 = predsDF2[['team_1', 'team_0','actual', 'predictions', 'truepos',
       'trueneg', 'falsepos', 'falseneg']]
predsDF3
predsDF3['precision'] = predsDF3['truepos']/(predsDF3['truepos']+predsDF3['falsepos'])
predsDF3['accuracy'] = (predsDF3['truepos']+predsDF3['trueneg'])/(predsDF3['truepos']+predsDF3['falsepos']+predsDF3['trueneg']+predsDF3['falseneg'])
predsDF3['recall'] = predsDF3['truepos']/(predsDF3['truepos']+predsDF3['falseneg'])
predsDF3 = predsDF3.reset_index()
predsDF3
```

```python
print("Average accuracy of the model:",predsDF3.accuracy.mean())
print("Average precision of the model:",predsDF3.precision.mean())
print("Average recall of the model:",predsDF3.recall.mean())
```
